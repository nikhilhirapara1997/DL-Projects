```python
##import Necessary Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import tensorflow as tf
```

```python
##import dataset
df= pd.read_csv('D:/Data Science/Python/Deep Learning/Churn Modelingg.csv')
```

```python
df
```

```python
##now drop the unnecessary data
df.drop(['RowNumber',"CustomerId","Surname"],axis=1,inplace=True)
```

```python
df
```

```python
##now splitting dataset into x & y
x=df.drop(["Exited"],axis=1)
y=df["Exited"]
```

```python
##for Categoricaldata we need to use encoding
geography=pd.get_dummies(x["Geography"],dtype=int,drop_first=True)
gender=pd.get_dummies(x["Gender"],dtype=int,drop_first=True)
```

```python
##drop categorical features
x.drop(["Geography","Gender"],axis=1,inplace=True)
```

```python
##concat the encoded features with main dataset
x=pd.concat([x,geography,gender],axis=1)
```

```python
## Train test Splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
## feature Scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
##Checking the shape of splitted datasets
print(x_train.shape)
y_train.shape
```

```python
##import model and layers from tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
##model
classifier=Sequential()
```

```python
#adding the input Layer
classifier.add(Dense(units=11,activation='ReLU'))
```

```python
## adding first hidden layer
classifier.add(Dense(units=7,activation='ReLU'))
```

```python
## adding second hidden layer
classifier.add(Dense(units=6,activation='ReLU'))
```

```python
##Output layer
classifier.add(Dense(1,activation='sigmoid'))
```

```python
##Early stopping
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
##need compile the model
classifier.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
##now fit the model on training dataset
model=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=100,callbacks=early_stopping)
```

```python
#predicting the test set results
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
#confusion metrics
from sklearn.metrics import confusion_matrix
cn=confusion_matrix(y_test,y_pred)
cn
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```
